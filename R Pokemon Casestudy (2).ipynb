```r
colnames(pokemon)[colnames(pokemon)=="Type.1"]<- "Primary_type" # I have changed height_nm column name to height
```

```r
#Now I want to see the data of only those Pokemon who's primary type is "Grass"
#so i will use the filter function.
pokemon %>% filter(Primary_type=="Grass")-> grass_pokemon
grass_pokemon
```

```r
#now we will visualize the data set for more understanding.
# we will try to figure out how grass type Pokemon are allocated to different generation.
ggplot(data = grass_pokemon, aes(x= Generation, fill= as.factor(Generation))) + geom_bar()
```

```r
#Now I want to plot a histogram to find out HP distribution of Grass type Pokemon

ggplot(data = grass_pokemon,aes(x=HP))+geom_histogram(fill="coral")
```

```r
#Now I want to see the data of only those Pokemon who's primary type is "Grass"
#so i will use the filter function.
pokemon %>% filter(Primary_type=="Fire")-> fire_pokemon
fire_pokemon
```

```r
colnames(pokemon)[colnames(pokemon)=="Sp..Atk"]<- "Special_Attack"  #Now we will change the column name in our data set
```

```r
colnames(pokemon)[colnames(pokemon)=="Sp..Def"]<- "Special_Defence" 
```

```r
colnames(pokemon)   #here we aill check whether the colomn name has been changed or not
```

```r
#Now I want to find out the correlation between sp_attack and sp_defence of fire type of Pokemon

ggplot(data=fire_pokemon,aes(x=Special_Attack,y=Special_Defence))+geom_point(col="hotpink")
```
